```python
from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = load_boston()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.columns = df.feature_names
```

```python
dataset
```

```python
dataset["Price"] = df.target
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1] # independent 
y = dataset.iloc[:,-1] # dependent
```

## Linear regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_regressor=LinearRegression()
mse=cross_val_score(lin_regressor,X,y,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

## lasso regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X,y)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

## Multivariate Analysis

```python
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,4,1)
df['sepal_length'].plot(kind='hist',color='olive',title='sepal length')
plt.subplot(1,4,2)
df['sepal_width'].plot(kind='hist',color='olive',title='sepal width')
plt.subplot(1,4,3)
df['petal_length'].plot(kind='hist',color='olive',title='Petal length')
plt.subplot(1,4,4)
df['petal_width'].plot(kind='hist',color='olive',title='petal length')
```

## Bivariate analysis

```python
x= df['sepal_length']
y= df['petal_length']
plt.scatter(x,y,color='skyblue')
```

```python
sns.scatterplot(data=df,x='sepal_length',y='petal_length',hue='species',size='species')
```

## Univariate Analysis

```python
df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal length')
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="species",size=3)
```
